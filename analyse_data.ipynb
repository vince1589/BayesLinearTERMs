```python
import pandas as pd
import glob
```

```python
ResFile = glob.glob("./results/*.csv")
dfs = []
for f in ResFile:
    df = pd.read_csv(f, header=None,index_col=0,skiprows=1).transpose()
    dfs.append(df)
```

```python
DF = pd.concat(dfs)
```

```python
DF.sort_values(by="MeanCor",ascending=False).head()
```

```python
DF.groupby(by=["File","RecepToRun","Poly","Prior"]).max()["MeanCor"]
```
